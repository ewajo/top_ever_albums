```python
import pandas as pd

dfs = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/',header=0)
type(dfs)
df=dfs[0]
df.rename(columns={"POS":"POZ","TITLE":"TYTUŁ","ARTIST":"ARTYSTA","YEAR":"ROK", "HIGH POSN":"MAX POZ"},inplace=True)
print(df)
unique_1=df['ARTYSTA'].unique()
print(unique_1)
print(len(unique_1))
vc=df['ARTYSTA'].value_counts()
print(vc) 
print('zespoły najczęściej występujące: Coldplay, Take that')
df.columns.str.title()
df.drop('MAX POZ', axis=1,inplace=True)
df.groupby('ROK').count()
df['ROK'] = pd.to_numeric(df.ROK)
SUMA=df[(df['ROK'] >- 1960) & (df['ROK'] <= 1990)].groupby('ROK').POZ.count()
print(SUMA)
print("Liczba albumów wydanych po 1960 i przed 1991 rokiem:", SUMA.sum())
print('Najmłodszy album wydano w roku:',df.ROK.min())
df.groupby(['ARTYSTA']).ROK.min()
writer = pd.ExcelWriter('pandas1.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='my_data1')
```
